```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
def main():
    
    df = None

    while True:
        print("\n--- Data Analysis Program ---")
        print("1. Load Data")
        print("2. Explore Data (Head)")
        print("3. Handle Missing Data")
        print("4. Summary Statistics (Describe)")
        print("5. Data Visualization")
        print("6. Save Last Visualization")
        print("7. Specific Column Statistics")
        print("8. Exit")
        
        choice = input("Select (1-8): ")

        if choice == '1':
            file = input("Enter file name (e.g., Covid.csv): ")
            try:
                df = pd.read_csv('Covid.csv')
                print(f"Successfully loaded {file}")
            except Exception as e:
                print(f"Error loading file: {e}")

        elif df is None and choice in ['2', '3', '4', '5', '6', '7']:
            print("Please load a dataset first (Option 1).")

        elif choice == '2':
            print(df.head())

        elif choice == '3':
            print("Rows with missing values:")
            print(df[df.isnull().any(axis=1)])

        elif choice == '4':
            print(df.describe())

        elif choice == '5':
            t = input("1. Bar 2. Line 3. Scatter: ")
            x = input("x Column name: ")
            y = input("y Column name: ")
            
            plt.figure(figsize=(10, 6))
            if t == '1':
                plt.bar(df[x], df[y])
            elif t == '2':
                plt.plot(df[x], df[y])
            elif t == '3':
                plt.scatter(df[x], df[y])
            
            plt.xlabel(x)
            plt.ylabel(y)
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()

        elif choice == '6':
            name = input("Save as (e.g., plot.png): ")
            plt.savefig(name)
            print(f"Saved as {name}")

        elif choice == '7':
            col = input("Enter the column name: ")
            
            print("\n--- Statistics ---")
            print("Mean: ", df[col].mean())
            print("Median: ", df[col].median())
            print("Mode: ", df[col].mode()[0]) # [0] gets the first mode
            print("Std Dev: ", df[col].std())
            print("Variance: ", df[col].var())
            print("\nValue Counts:\n", df[col].value_counts())
            print("Column not found!")

        elif choice == '8':
            print("Exiting...")
            break
        else:
            print("Invalid choice, try again.")

if __name__ == "__main__":
    main()
```
